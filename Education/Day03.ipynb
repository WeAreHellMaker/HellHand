## 1. 만약에... (if)

파이썬에서 가장 기본적인 조건문은 `if`입니다. 영어 뜻 그대로 "만약 ~라면"이라는 뜻이죠.

```python
if 조건:
    실행할_명령  # 조건이 참(True)일 때만 실행됩니다.
```

**중요!** 파이썬에서는 `if`문 다음 줄에 반드시 **들여쓰기(Tab 혹은 스페이스 4칸)**를 해야 합니다. 들여쓰기가 된 부분만 `if`의 영향을 받습니다.

````python
# 로봇의 배터리 상태를 체크해봅시다.
battery = 10

if battery < 20:
    print("경고! 배터리가 부족합니다.")
    print("충전해주세요.")

print("시스템 점검 끝.") # 이 줄은 들여쓰기가 없어서 항상 실행됩니다.
````

## 2. 그렇지 않으면... (else)

`if`의 조건이 맞지 않을 때 실행할 코드는 `else`에 적습니다.

```python
if 조건:
    참일_때_실행
else:
    거짓일_때_실행
```

````python
is_connected = False  # 로봇 연결 상태

if is_connected:
    print("로봇과 통신 중입니다.")
else:
    print("연결이 끊어졌습니다. 케이블을 확인하세요.")
````

## 3. 그게 아니고 만약... (elif)

조건이 여러 개일 때는 `elif` (else if의 줄임말)를 사용합니다.
가위바위보처럼 경우의 수가 3개 이상일 때 유용합니다.

````python
finger_count = 2  # 펴진 손가락 개수

if finger_count == 0:
    print("주먹!")
elif finger_count == 2:
    print("가위!")
elif finger_count == 5:
    print("보!")
else:
    print("알 수 없는 손모양입니다.")
````

### 3.1. 비교 연산자 정리
- `==`: 같다 (등호 2개!)
- `!=`: 다르다
- `>` , `<`: 크다, 작다
- `>=` , `<=`: 크거나 같다, 작거나 같다

## 4. 리스트와 조건문의 만남

지난 시간에 배운 리스트와 오늘 배운 조건문을 합치면 더 복잡한 로직을 만들 수 있습니다.
예를 들어, 엄지 손가락의 각도만 확인해볼까요?

````python
# [엄지, 검지, 중지, 약지, 소지] 각도 리스트
angles = [100, 0, 0, 0, 0]

# 엄지(0번)가 90도 이상 굽혀져 있는지 확인
if angles[0] >= 90:
    print("엄지가 굽혀졌습니다.")
else:
    print("엄지가 펴져 있습니다.")
````
