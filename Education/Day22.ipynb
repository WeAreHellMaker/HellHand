"가위바위보"를 인식해서, 로봇이 이기는 수를 내도록 만들어볼까요?

## 1. 가위, 바위, 보 판단하기

각 손가락이 펴졌는지(1), 굽혀졌는지(0)를 기준으로 판단합니다.
보통 **100도 이상**이면 펴진 것으로 봅니다.

- **바위**: 모든 손가락 0
- **보**: 모든 손가락 1
- **가위**: 검지, 중지만 1

```python
def detect_gesture(angles):
    # angles: [엄지, 검지, 중지, 약지, 소지] 각도 리스트
    
    # 1. 펴짐/굽힘 상태로 변환 (Threshold: 100도)
    state = []
    for a in angles:
        if a >= 100: state.append(1) # 펴짐
        else: state.append(0) # 굽힘
    
    # 2. 패턴 매칭
    if state == [0, 0, 0, 0, 0]: return "ROCK"
    if state == [1, 1, 1, 1, 1]: return "PAPER"
    if state == [0, 1, 1, 0, 0]: return "SCISSORS"
    if state == [1, 1, 0, 0, 1]: return "SPIDERMAN"
    
    return "UNKNOWN"
```

## 2. 절대 승리 로봇 만들기

내가 '가위'를 내면 로봇은 '바위'를 내도록 코드를 짜봅시다.

```python
def get_winning_move(user_gesture):
    if user_gesture == "SCISSORS":
        return [0, 0, 0, 0, 0] # 바위 (다 굽힘)
    elif user_gesture == "ROCK":
        return [180, 180, 180, 180, 180] # 보 (다 폅)
    elif user_gesture == "PAPER":
        return [0, 180, 180, 0, 0] # 가위
    else:
        return None
```
