## 1. 하나씩 꺼내보기 (For 문)

`for`문은 리스트 안에 있는 물건들을 하나씩 순서대로 꺼내서 작업을 처리할 때 씁니다.

```python
for 변수 in 리스트:
    반복할_명령
```

````python
fingers = ["엄지", "검지", "중지", "약지", "소지"]

for finger in fingers:
    print(finger + " 손가락을 움직입니다.")
    
print("모든 손가락 점검 완료!")
````

### 1.1. 숫자 범위로 반복하기 (range)

리스트가 없어도, "5번 반복해!"라고 시킬 수 있습니다. 이때 `range()` 함수를 씁니다.

- `range(5)`: 0부터 4까지 (총 5개) 숫자를 만들어줍니다. **(끝 숫자는 포함 안 됨!)**

````python
for i in range(5):
    print(i, "번째 모터 테스트 중...")
````

## 2. 조건이 맞을 때까지 반복 (While 문)

`for`문이 "몇 번" 반복할지 정해져 있다면, `while`문은 "~할 때까지" 반복합니다.

```python
while 조건:
    반복할_명령
```

**주의!** `while`문을 쓸 때는 조건이 언젠가는 거짓(False)이 되어 끝나도록 해야 합니다. 그렇지 않으면 영원히 멈추지 않는 **무한 루프**에 빠집니다.

````python
battery = 20

while battery > 0:
    print("로봇 작동 중... 배터리 잔량:", battery)
    battery = battery - 5  # 배터리를 5씩 씁니다.
    
print("배터리 방전! 로봇이 멈췄습니다.")
````

## 3. 반복문 제어하기 (break, continue)

- **`break`**: 반복문을 강제로 탈출합니다. (비상 정지 버튼!)
- **`continue`**: 이번 순서만 건너뛰고 다음 순서로 넘어갑니다.

````python
# 위험한 각도(180도 이상)가 발견되면 즉시 검사를 중단해봅시다.
angles = [50, 60, 200, 70, 80]

for angle in angles:
    if angle >= 180:
        print("위험!", angle, "도 발견! 시스템 긴급 정지!")
        break
    print(angle, "도 - 정상")
````
