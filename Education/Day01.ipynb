## 2. Jupyter Notebook 다루기

이 화면은 **Jupyter Notebook**입니다. 마치 공책(Notebook)처럼 설명글과 코드를 자유롭게 적고 실행할 수 있는 도구죠.
데이터 분석가나 AI 개발자들이 가장 사랑하는 도구이기도 합니다.

### 2.1. 셀(Cell)의 종류
- **마크다운(Markdown) 셀**: 지금 읽고 계신 설명글입니다. 코드로 실행되지 않습니다.
- **코드(Code) 셀**: `In [ ]:` 표시가 있는 회색 박스입니다. 실제로 파이썬 코드가 실행되는 곳입니다.

### 2.2. 필수 단축키 (이것만은 꼭!)
- **`Shift + Enter`**: 현재 셀을 실행하고 다음 셀로 넘어갑니다. (가장 많이 씁니다!)
- **`Ctrl + Enter`**: 현재 셀을 실행하고 제자리에 머뭅니다.
- **`A`** (셀 선택 상태에서): 현재 셀 **위(Above)**에 새로운 셀을 추가합니다.
- **`B`** (셀 선택 상태에서): 현재 셀 **아래(Below)**에 새로운 셀을 추가합니다.
- **`D`, `D`** (셀 선택 상태에서 D를 두 번): 현재 셀을 **삭제(Delete)**합니다.
