## 1. 변수 (Variable): 하나의 값을 담는 그릇

변수는 어떤 값을 저장해두고, 나중에 꺼내 쓸 수 있도록 이름을 붙여준 '그릇'과 같습니다.

### 1.1. 변수 만들기 (할당)
파이썬에서는 `이름 = 값` 형태로 변수를 만듭니다. (수학에서 '같다'가 아니라, 오른쪽의 값을 왼쪽 그릇에 '담는다'는 뜻입니다!)

```python
# 로봇의 정보를 변수에 담아봅시다.
robot_name = "Hell Hand"   # 문자열(글자) 저장
thumb_angle = 90           # 정수(숫자) 저장
battery_level = 85.5       # 실수(소수점 숫자) 저장

print(robot_name)
print(thumb_angle)
print(battery_level)
```

### 1.2. 변수 이름 짓기 규칙 (Naming Convention)
변수 이름은 마음대로 지을 수 있지만, 개발자들 사이의 약속이 있습니다.

1. **알아보기 쉽게 짓기**: `a = 90` 보다는 `thumb_angle = 90`이 훨씬 이해하기 쉽겠죠?
2. **Under bar**: 단어 사이를 밑줄(`_`)로 연결합니다. (예: `my_robot_hand`)
3. **숫자로 시작 금지**: `1st_finger` (X) -> `first_finger` (O)
4. **특수문자 금지**: 띄어쓰기나 `!`, `@` 등은 쓸 수 없습니다.

## 2. 리스트 (List): 여러 값을 한꺼번에 담는 기차

우리가 만들 로봇 손에는 손가락이 5개나 있습니다.
`thumb`, `index`, `middle`, `ring`, `pinky`... 변수를 5개나 만드는 건 너무 귀찮죠.

이럴 때 **리스트**를 사용하면 관련된 여러 데이터를 한 묶음으로 관리할 수 있습니다.

```python
# 손가락 5개의 각도를 리스트 하나에 저장해봅시다.
# 순서대로 [엄지, 검지, 중지, 약지, 소지] 입니다.
finger_angles = [0, 10, 20, 30, 40]

print("모든 손가락 각도:", finger_angles)
```

### 2.1. 인덱싱 (Indexing): 리스트에서 값 꺼내기
리스트 안에 있는 특정 값만 꺼내고 싶다면 **번호표(Index)**를 사용합니다.
**주의! 파이썬의 번호는 0번부터 시작합니다.**

- 0번: 엄지
- 1번: 검지
- 2번: 중지
- 3번: 약지
- 4번: 소지

```python
# 엄지 손가락(0번)의 각도만 보고 싶다면?
print("엄지 각도:", finger_angles[0])

# 중지 손가락(2번)의 각도만 보고 싶다면?
print("중지 각도:", finger_angles[2])
```

### 2.2. 리스트 값 바꾸기
로봇이 움직이면 각도가 변하겠죠? 리스트의 값은 언제든 바꿀 수 있습니다.

```python
print("변경 전:", finger_angles)

# 검지(1번)를 90도로 굽혀라!
finger_angles[1] = 90

print("변경 후:", finger_angles)
```
